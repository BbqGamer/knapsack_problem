```python
from matplotlib import pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 14,
                     'figure.dpi': 150,
                    })
```

```python
df = pd.read_csv("../output/run1.csv")
pd.pivot_table(df, values="time", index="n", columns="method", aggfunc="mean").plot(
    title="Time to solve the Knapsack Problem",
    ylabel="Time (ms)",
    figsize=(8, 5),
    grid=True
)
```

```python
df = pd.read_csv("../output/run1.csv")
pd.pivot_table(df, values="time", index="n", columns="method", aggfunc="mean").plot(
    title="Time to solve the Knapsack Problem (logarithmic scale)",
    ylabel="Time (ms)",
    figsize=(8, 5),
    logy = True,
    grid = True,
)
```

```python
pd.pivot_table(df[df.method.isin(["Dynamic", "Greedy"])], values="time", index="n", columns="method", aggfunc="mean").plot(
    title="Time to solve the Knapsack Problem (Dynamic vs Greedy)",
    ylabel="Time (ms)",
    figsize=(8, 5),
    grid=True
)
```

```python
pd.pivot_table(df[df.method.isin(["Dynamic", "Greedy"])], values="time", index="n", columns="method", aggfunc="mean").plot(
    title="Time to solve the Knapsack Problem - log scale",
    logy = True,
    ylabel="Time (ms)",
    grid = True,
    figsize=(8, 5),
)
```

```python
df = pd.read_csv("../output/run1.csv")
pd.pivot_table(df[df.method.isin(["Dynamic", "Greedy"])], values="result", index="n", columns="method", aggfunc="max").plot(
    title="Time to solve the Knapsack Problem (logarithmic scale)",
    ylabel="Time (ms)",
    figsize=(8, 5),
    grid = True,
)
```
